### Download String-DB PPI files for each species in AGE Datasets.

```python
import pandas as pd

# https://stringdb-downloads.org/download/protein.links.v12.0/48.protein.links.v12.0.txt.gz

# requests.get(f"https://stringdb-downloads.org/download/protein.links.v12.0/{taxon_id}.protein.links.v12.0.txt.gz")


# Turns the tsv files into csv files
df_string = pd.read_csv('data/string_species.csv')
df_age = pd.read_csv('data/age_species.csv')
```

```python
df_to_append = pd.DataFrame(columns=df_string.columns)

for i, df in df_age.iterrows():
    genus = df["Genus"]
    species = df["Species"]
    combined = f"{genus} {species}"
    if combined in df_string["official_name_NCBI"].values:
        print("Found one!")
        row_to_append = df_string[df_string['official_name_NCBI'] == combined]
        df_to_append = pd.concat([df_to_append, row_to_append], ignore_index=True)




    
```

```python
df_to_append.to_csv("data/mapped_species.csv")
```
